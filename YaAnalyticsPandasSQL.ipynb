Переведем даты в `timestamp`.

Количество времени потраченное на прослушивание каждого трека в секундах.

```python
df['listen_times_sec'] = (df['utc_audition_end_dttm'] - df['utc_audition_start_dttm']).astype('timedelta64[s]')
```

Создадим новые перменные:
* `track_complete` - прослушан ли трек до конца;
* `track_complete_ratio` - доля прослушанного трека.

```python
df['track_complete'] = (df['listen_times_sec'] == df['track_duration']) * 1
```

```python
df['track_complete_ratio'] = df['listen_times_sec'] / df['track_duration']
```

Посмотрим на жанры.

```python
df['track_genre'].value_counts()
```

Количество уникальных юзеров, слушавших хотя бы раз жанр 'TECHNO'.

```python
df[df['track_genre']=='TECHNO']['user_id'].nunique()
```

Количество юзеров, хотя бы раз слушавших 'TECHNO' равно количеству уникальных юзеров в датасете - все юзеры из выборки слушали хотя бы раз жанр 'TECHNO'.

```python
df[df['track_genre']=='TECHNO']['user_id'].nunique() == df['user_id'].nunique()
```

Добавим таким юзерам в отдельной колонке маркер.

```python
df['is_techno'] = (df['track_genre']=='TECHNO') * 1
```

Посчитаем долю треков жанра 'TECHNO' от всех треков для каждого юзера и другие полезные признаки.

```python
techno_frac = df.groupby('user_id')[['is_techno', 
                       'track_id', 
                       'utc_audition_start_dttm', 
                       'track_complete_ratio']].agg({
    'is_techno': ['count', 'mean'], # 'count' - количество прослушанных треков, всего; 'mean' - доля треков в жанре 'TECHNO'
    'track_id': ['nunique'], # Количество уникальных треков которые прослушал юзер
    'utc_audition_start_dttm': ['max'], # Когда юзер последний раз слушал
    'track_complete_ratio': ['mean'] # Среднее время дослушивания трека
})
```

Уберем "двухэтажные" индексы в колонка новой таблицы и переименуем колонки.

```python
techno_frac.columns = techno_frac.columns.get_level_values(0)
```

```python
techno_frac.columns = ['tracks_count', 
                       'techno_ratio', 
                       'unique_tracks_count', 
                       'last_dt', 
                       'track_complete_ratio_mean']
```

Доля уникалных треков от общего числа прослушанных треков.

```python
techno_frac['unique_ratio'] = techno_frac['unique_tracks_count'] / techno_frac['tracks_count']
```

Переведем `last_dt` из `timestamp` в `unix` (секунды), чтобы удобнее было соритировать.

```python
techno_frac['last_dt'] = techno_frac['last_dt'].map(pd.Timestamp.timestamp)
```

* `tracks_count` - количество прослушанных юзером треков, всего. Чем больше - тем лучше.
* `techno_ratio` - доля треков в жанре 'TECHNO' от общего количества прослушанных юзером треков. Чем больше - тем лучше.
* `unique_tracks_count` - количество уникальных треков, всего. Чем больше - тем лучше.
* `last_dt` - дата послднего прослушивания в секундах. Чем больше - тем позднее (лучше).
* `track_complete_ratio_mean` - доля прослушанных до конца треков. Чем больше - тем лучше.
* `unique_ratio` - доля уникальных треков от общего количества прослушанных треков. Чем больше - тем лучше.

**RFM**

```python
techno_frac['tracks_count'].rank(pct=True)
```

Посчитаем такие скоры для всех переменных.

`method='max'` - при одинаковых значениях в группе - будет возвращать максимальное значение группы.

`pct=True` - отображение рейтнга в процентилях.

```python
features = techno_frac.columns.to_list()

for f in features:
    print(f)
    techno_frac[f'{f}_pct'] = techno_frac[f].rank(pct=True, method='max')
```

Сложим скоры по отдельным переменным в единую и отсортируем юзеров, получив, таким образом, наиболее подходящих - тех, которые набрали максимальное количество скоров по каждой из переменных.

```python
techno_frac['score'] = techno_frac[['tracks_count_pct', 
                                    'techno_ratio_pct', 
                                    'unique_tracks_count_pct', 
                                    'last_dt_pct', 
                                    'track_complete_ratio_mean_pct', 
                                    'unique_ratio_pct']].sum(axis=1)
```

```python
techno_frac.sort_values('score', ascending=False)
```

В качестве дополнения, можно поиграть с весами фичей, например отдать больший вес `techno_ratio`.

```python
techno_frac['score_weighted'] = (0.16/6 * techno_frac['tracks_count_pct'] + \
                        2/6 * techno_frac['techno_ratio_pct'] + \
                        0.16/6 * techno_frac['unique_tracks_count_pct'] + \
                        0.16/6 * techno_frac['last_dt_pct'] + \
                        0.16/6 * techno_frac['track_complete_ratio_mean_pct'] + \
                        0.16/6 * techno_frac['unique_ratio_pct'])
```

Сделаем функцию, которая будет читать SQL-запрос и возвращать pandas-датфрейм.

```python
def select(sql):
    return pd.read_sql(sql, con)
```

Джойним таблицы.

Так как нас интересуют юзеры и их покупки только в период действия подписки "Яндекс Плюс", то джойним по `user_id` и условию, что покупки были между `utc_dttm_from` и `utc_dttm_to`.

```python
sql = '''SELECT *

FROM market_orders m

LEFT JOIN active_plus a
ON m.user_id = a.user_id AND
date(m.utc_creation_dttm) BETWEEN date(a.utc_dttm_from) AND date(a.utc_dttm_to)'''
```

```python
select(sql)
```

Проставим всем юзерам напротив тех покупок, когда была активна подписка флаг $1$, если не активна - $0$.

```python
sql = '''SELECT m.user_id,
CASE WHEN a.user_id IS NOT NULL THEN 1 ELSE 0 END as active_plus_enable

FROM market_orders m

LEFT JOIN active_plus a
ON m.user_id = a.user_id AND
date(m.utc_creation_dttm) BETWEEN date(a.utc_dttm_from) AND date(a.utc_dttm_to)'''
```

```python
select(sql)
```

Посчитаем среднее количество покупок с активной подпиской на каждого юзера.

```python
sql = '''SELECT m.user_id,
AVG(CASE WHEN a.user_id IS NOT NULL THEN 1 ELSE 0 END) as active_plus_enable_mean

FROM market_orders m

LEFT JOIN active_plus a
ON m.user_id = a.user_id AND
date(m.utc_creation_dttm) BETWEEN date(a.utc_dttm_from) AND date(a.utc_dttm_to)

GROUP BY m.user_id'''
```

```python
select(sql)
```

Оставим только тех, у кого доля таких покупок $>= 80\%$.

```python
sql = '''SELECT m.user_id,
AVG(CASE WHEN a.user_id IS NOT NULL THEN 1 ELSE 0 END) as active_plus_enable_mean

FROM market_orders m

LEFT JOIN active_plus a
ON m.user_id = a.user_id AND
date(m.utc_creation_dttm) BETWEEN date(a.utc_dttm_from) AND date(a.utc_dttm_to)

GROUP BY m.user_id

HAVING AVG(CASE WHEN a.user_id IS NOT NULL THEN 1 ELSE 0 END) >= 0.8'''
```

```python
select(sql)
```

Посчитаем количество юзеров.

```python
sql = '''WITH temp AS (
SELECT m.user_id,
AVG(CASE WHEN a.user_id IS NOT NULL THEN 1 ELSE 0 END) as active_plus_enable_mean

FROM market_orders m

LEFT JOIN active_plus a
ON m.user_id = a.user_id AND
date(m.utc_creation_dttm) BETWEEN date(a.utc_dttm_from) AND date(a.utc_dttm_to)

GROUP BY m.user_id

HAVING AVG(CASE WHEN a.user_id IS NOT NULL THEN 1 ELSE 0 END) >= 0.8)



SELECT COUNT(user_id) as active_users_with_plus
FROM temp'''
```
